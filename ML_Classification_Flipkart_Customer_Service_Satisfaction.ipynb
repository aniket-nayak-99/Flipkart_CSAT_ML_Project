## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

df_raw = df.copy()   # keeping original copy for safety


# -------------------------------------------------------
# ‚úÖ 1. Clean Column Names (remove spaces, special chars)
# -------------------------------------------------------
df.columns = df.columns.str.strip().str.replace(" ", "_").str.replace("-", "_")


# -------------------------------------------------------
# ‚úÖ 2. Convert Date & Time Columns to datetime format
# -------------------------------------------------------
date_cols = [
    "order_date_time",
    "Issue_reported_at",
    "issue_responded",
    "Survey_response_Date"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")


# -------------------------------------------------------
# ‚úÖ 3. Create Time-Based Features
# -------------------------------------------------------

# Response time: time taken by agent to respond
df["response_time_minutes"] = (
    (df["issue_responded"] - df["Issue_reported_at"])
    .dt.total_seconds() / 60
)

# Time gap between order and issue raised
df["order_to_issue_hours"] = (
    (df["Issue_reported_at"] - df["order_date_time"])
    .dt.total_seconds() / 3600
)

# Time gap between response and survey
df["response_to_survey_hours"] = (
    (df["Survey_response_Date"] - df["issue_responded"])
    .dt.total_seconds() / 3600
)


# -------------------------------------------------------
# ‚úÖ 4. Convert numerical columns to correct dtype
# -------------------------------------------------------
num_cols = ["Item_price", "connected_handling_time"]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")


# -------------------------------------------------------
# ‚úÖ 5. Strip and clean categorical text columns
# -------------------------------------------------------
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip()


# -------------------------------------------------------
# ‚úÖ 6. Handle Duplicates
# -------------------------------------------------------
print("Duplicates before cleaning:", df.duplicated().sum())
df.drop_duplicates(inplace=True)
print("Duplicates after cleaning:", df.duplicated().sum())


# -------------------------------------------------------
# ‚úÖ 7. Handle Missing Values (simple imputation)
# -------------------------------------------------------

# Fill missing numeric values with median
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill missing time-based engineered features
df[["response_time_minutes",
    "order_to_issue_hours",
    "response_to_survey_hours"]] = df[
    ["response_time_minutes",
     "order_to_issue_hours",
     "response_to_survey_hours"]
].fillna(0)

# Fill categorical missing with "Unknown"
df[cat_cols] = df[cat_cols].replace("nan", np.nan)
df[cat_cols] = df[cat_cols].fillna("Unknown")


# -------------------------------------------------------
# ‚úÖ 8. Display Final Clean Dataset Info
# -------------------------------------------------------
print("\n‚úÖ Data Wrangling Completed Successfully!")
df.info()
df.head()
````

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# ‚úÖ Statistical Test: One-Way ANOVA
# We test if mean response_time_minutes differs across CSAT groups

from scipy.stats import f_oneway

groups = [df[df['CSAT_Score'] == score]['response_time_minutes'].dropna()
          for score in sorted(df['CSAT_Score'].unique())]

anova_stat, p_value = f_oneway(*groups)

anova_stat, p_value
````

I performed a **One-Way ANOVA (Analysis of Variance)** test.

- The independent variable (**CSAT Score**) is **categorical with multiple classes**.

### 4. Textual Data Preprocessing

#### 1. Expand Contraction

````python
# -----------------------------
# ‚úÖ Text Preprocessing: Expand Contractions
# -----------------------------

import re

# Contraction dictionary (common English contractions)
contraction_dict = {
    "can't": "cannot",
    "won't": "will not",
    "n't": " not",
    "'re": " are",
    "'s": " is",
    "'d": " would",
    "'ll": " will",
    "'t": " not",
    "'ve": " have",
    "'m": " am",
}

# Function to expand contractions
def expand_contractions(text):
    if isinstance(text, str):
        for contraction, expansion in contraction_dict.items():
            text = re.sub(contraction, expansion, text)
        return text
    return text

# Apply to the textual column(s)
text_cols = ["Customer_Remarks"]

for col in text_cols:
    df[col] = df[col].apply(expand_contractions)

print("‚úÖ Contractions Expanded in Text Columns:", text_cols)
df[text_cols].head()
````

#### 2. Lower Casing

````python
# -----------------------------
# ‚úÖ Text Preprocessing: Lower Casing
# -----------------------------

# Apply lowercase transformation to text-based columns
text_cols = ["Customer_Remarks"]

for col in text_cols:
    df[col] = df[col].astype(str).str.lower()

print("‚úÖ Lowercasing Applied to Text Columns:", text_cols)
df[text_cols].head()
````

#### 3. Removing Punctuations

````python
# -----------------------------------------
# ‚úÖ Text Preprocessing: Remove Punctuations
# -----------------------------------------

import string

text_cols = ["Customer_Remarks"]

# Function to remove punctuation
def remove_punct(text):
    if isinstance(text, str):
        return text.translate(str.maketrans("", "", string.punctuation))
    return text

# Apply to all text columns
for col in text_cols:
    df[col] = df[col].apply(remove_punct)

print("‚úÖ Punctuations Removed from:", text_cols)
df[text_cols].head()
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# ----------------------------------------------------------------
# ‚úÖ Text Preprocessing: Remove URLs & Words Containing Digits
# ----------------------------------------------------------------

import re

def remove_urls_and_digits(text):
    if isinstance(text, str):

        # Remove URLs (http, https, www)
        text = re.sub(r"http\S+|www\S+|https\S+", "", text)

        # Remove words containing digits (e.g., "abc123", "12days")
        text = re.sub(r"\w*\d\w*", "", text)

        return text.strip()
    return text

# Apply to text columns
for col in text_cols:
    df[col] = df[col].apply(remove_urls_and_digits)

print("‚úÖ URLs & Digit-containing words removed from:", text_cols)
df[text_cols].head()
````

#### 5. Removing Stopwords & Removing White spaces

````python
# -----------------------------------------
# ‚úÖ Text Preprocessing: Remove Stopwords
# -----------------------------------------

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

text_cols = ["Customer_Remarks"]

def remove_stopwords(text):
    if isinstance(text, str):
        return " ".join([word for word in text.split() if word not in stop_words])
    return text

for col in text_cols:
    df[col] = df[col].apply(remove_stopwords)

print("‚úÖ Stopwords Removed from:", text_cols)
df[text_cols].head()
````

````python
# -----------------------------------------
# ‚úÖ Text Preprocessing: Remove Extra White Spaces
# -----------------------------------------

def remove_extra_spaces(text):
    if isinstance(text, str):
        # Replace multiple spaces with a single space
        return re.sub(r'\s+', ' ', text).strip()
    return text

for col in text_cols:
    df[col] = df[col].apply(remove_extra_spaces)

print("‚úÖ Extra White Spaces Removed from:", text_cols)
df[text_cols].head()
````

#### 6. Rephrase Text

````python
# ---------------------------------------------------------
# ‚úÖ Text Preprocessing: Rephrase / Normalize Text Phrases
# ---------------------------------------------------------

# Dictionary to normalize common phrases in customer remarks
rephrase_dict = {
    "cant": "cannot",
    "wont": "will not",
    "didnt": "did not",
    "dont": "do not",
    "pls": "please",
    "plz": "please",
    "asap": "as soon as possible",
    "u": "you",
    "ur": "your",
    "msg": "message",
    "cust": "customer",
    "info": "information",
    "abt": "about",
    "btw": "by the way",
    "idk": "i do not know",
}

def rephrase_text(text):
    if isinstance(text, str):
        words = text.split()
        rephrased_words = [rephrase_dict.get(w, w) for w in words]
        return " ".join(rephrased_words)
    return text

# Apply to text columns
text_cols = ["Customer_Remarks"]
for col in text_cols:
    df[col] = df[col].apply(rephrase_text)

print("‚úÖ Rephrasing Completed for:", text_cols)
df[text_cols].head()
````

#### 7. Tokenization

````python
# -----------------------------------------
# ‚úÖ Text Preprocessing: Tokenization
# -----------------------------------------

import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Download missing resource
from nltk.tokenize import word_tokenize

text_cols = ["Customer_Remarks"]

def tokenize_text(text):
    if isinstance(text, str):
        return word_tokenize(text)
    return text

# Apply tokenization
for col in text_cols:
    df[col] = df[col].apply(tokenize_text)

print("‚úÖ Tokenization applied to:", text_cols)
df[text_cols].head()
````

#### 8. Text Normalization

````python
# ------------------------------------------------------
# ‚úÖ Text Normalization (Lemmatization)
# ------------------------------------------------------

import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# Lemmatization function
def lemmatize_text(tokens):
    if isinstance(tokens, list):
        return [lemmatizer.lemmatize(word) for word in tokens]
    elif isinstance(tokens, str):
        # If text wasn't tokenized earlier
        return " ".join([lemmatizer.lemmatize(word) for word in tokens.split()])
    return tokens

# Apply to tokenized text column
text_cols = ["Customer_Remarks"]

for col in text_cols:
    df[col] = df[col].apply(lemmatize_text)

print("‚úÖ Lemmatization Applied Successfully on:", text_cols)
df[text_cols].head()
````

### ‚úÖ Which text normalization technique have you used and why?

**Benefits:**
- Reduces vocabulary size  
- Improves model generalization  
- Ensures text consistency for vectorization (TF-IDF, Word2Vec, etc.)

‚úÖ In this project, Lemmatization was used because it provides cleaner, meaningful text that enhances NLP feature extraction and model performance.

````python
# ---------------------------------------------------------
# ‚úÖ Feature Manipulation: Reduce correlation & create new features
# ---------------------------------------------------------

import numpy as np
import pandas as pd

df_feat = df.copy()

# ---------------------------------------------------------
# ‚úÖ 1. Log-Transform Skewed Features
# ---------------------------------------------------------

skewed_cols = ["response_time_minutes", "order_to_issue_hours",
               "response_to_survey_hours", "connected_handling_time"]

for col in skewed_cols:
    df_feat[col] = df_feat[col].clip(lower=0)  # prevent -inf
    df_feat[col + "_log"] = np.log1p(df_feat[col])

print("‚úÖ Log-transform applied")


# ---------------------------------------------------------
# ‚úÖ 2. Interaction Features
# ---------------------------------------------------------

df_feat["handling_response_ratio"] = (
    df_feat["connected_handling_time"] / (df_feat["response_time_minutes"] + 1)
)

df_feat["total_resolution_minutes"] = (
    df_feat["response_time_minutes"]
    + df_feat["order_to_issue_hours"] * 60
    + df_feat["response_to_survey_hours"] * 60
)

print("‚úÖ Interaction features created")


# ---------------------------------------------------------
# ‚úÖ 3. Safe Binning
# ---------------------------------------------------------

def safe_bin(col, bins=4):
    try:
        return pd.qcut(df_feat[col], q=bins, labels=False, duplicates="drop") + 1
    except:
        return pd.Series([1] * len(df_feat))

df_feat["price_bucket"] = safe_bin("Item_price")
df_feat["handling_bucket"] = safe_bin("connected_handling_time")

print("‚úÖ Binning applied safely")


# ---------------------------------------------------------
# ‚úÖ 4. Drop Highly Correlated Raw Columns
# ---------------------------------------------------------

df_feat.drop(columns=[
    "response_time_minutes",
    "order_to_issue_hours",
    "response_to_survey_hours"
], inplace=True)

print("‚úÖ Dropped correlated raw columns")


# ---------------------------------------------------------
# ‚úÖ Final Output
# ---------------------------------------------------------

print("\n‚úÖ Feature Manipulation Completed")
print("Final Shape:", df_feat.shape)
df_feat.head()
````

````python
# ---------------------------------------------------------
# ‚úÖ Feature Selection for CSAT Prediction
# ---------------------------------------------------------

from sklearn.feature_selection import mutual_info_classif, chi2, f_classif, SelectKBest
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

# Make a clean copy
df_fs = df_feat.copy()

# -----------------------------
# ‚úÖ Split into Features & Target
# -----------------------------
X = df_fs.drop("CSAT_Score", axis=1)
y = df_fs["CSAT_Score"]

# Ensure all features are numeric
X = X.select_dtypes(include=[np.number])

# =========================================================
# ‚úÖ 1. MUTUAL INFORMATION (Captures Non-Linear Relations)
# =========================================================

mi_selector = SelectKBest(mutual_info_classif, k=10)
mi_selector.fit(X, y)

mi_scores = pd.Series(mi_selector.scores_, index=X.columns)\
                .sort_values(ascending=False)


# =========================================================
# ‚úÖ 2. CHI-SQUARE TEST (For categorical/positive numeric data)
# =========================================================

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

chi_selector = SelectKBest(chi2, k=10)
chi_selector.fit(X_scaled, y)

chi_scores = pd.Series(chi_selector.scores_, index=X.columns)\
                .sort_values(ascending=False)


# =========================================================
# ‚úÖ 3. ANOVA F-test (Linear relationship strength)
# =========================================================

anova_selector = SelectKBest(f_classif, k=10)
anova_selector.fit(X, y)

anova_scores = pd.Series(anova_selector.scores_, index=X.columns)\
                .sort_values(ascending=False)


# =========================================================
# ‚úÖ 4. RANDOM FOREST FEATURE IMPORTANCE (Tree-based)
# =========================================================

rf = RandomForestClassifier(n_estimators=150, random_state=42)
rf.fit(X, y)

rf_scores = pd.Series(rf.feature_importances_, index=X.columns)\
                .sort_values(ascending=False)


# =========================================================
# ‚úÖ Display Results
# =========================================================

print("\n‚úÖ Top Features by Mutual Information:\n", mi_scores.head(10))
print("\n‚úÖ Top Features by Chi-Square:\n", chi_scores.head(10))
print("\n‚úÖ Top Features by ANOVA F-test:\n", anova_scores.head(10))
print("\n‚úÖ Top Features by Random Forest Importance:\n", rf_scores.head(10))


# Optional: combine all scores in one table
feature_rank_df = pd.DataFrame({
    "Mutual_Info": mi_scores,
    "Chi2": chi_scores,
    "ANOVA_F": anova_scores,
    "RF_Importance": rf_scores
}).fillna(0)

print("\n‚úÖ Combined Feature Ranking Table (Top 15):\n")
feature_rank_df.sort_values(by="RF_Importance", ascending=False).head(15)
````

### ‚úÖ Why Class Weights?

   SMOTE generates artificial samples, which can:
   - introduce noise  
   - distort real customer behavior  
   - cause overfitting in large datasets  

   Class weights avoid this.

2. **Works extremely well with tree-based models**  
   Random Forest, XGBoost, and Gradient Boosting naturally integrate class weights to:
   - penalize errors on minority classes  
   - reduce bias toward the dominant class (CSAT = 5)

3. **Dataset is already large (~86k rows)**  
   Oversampling would:
   - Increase training time  
   - Increase memory usage  
   - Produce diminishing returns

4. **Retains interpretability of real CSAT patterns**

5. **Balanced treatment of minority classes**  
   The model pays more attention to rare but important dissatisfied customers (CSAT 1, 2, 3).
